```python
from pathlib import Path
import json
import subprocess

REGION = "eu-west-1"
TRAINING_JOB_NAMES = [
    "src-train-f2llm-v2-160m-embedding-20260327071322",
    "src-train-gist-small-embedding-20260326170101",
    "src-train-minilm-embedding-20260326150308",
    "src-train-static-embedding-20260326150215",
]


def find_repo_root(start: Path) -> Path:
    for path in [start, *start.parents]:
        if (path / "pyproject.toml").exists():
            return path
    return start


repo_root = find_repo_root(Path.cwd())
out_dir = repo_root / "cost comparison data"
out_dir.mkdir(parents=True, exist_ok=True)

for job_name in TRAINING_JOB_NAMES:
    cmd = [
        "aws",
        "sagemaker",
        "describe-training-job",
        "--region",
        REGION,
        "--training-job-name",
        job_name,
        "--output",
        "json",
    ]

    result = subprocess.run(cmd, check=True, capture_output=True, text=True)
    payload = json.loads(result.stdout)

    out_file = out_dir / f"{job_name}.json"
    out_file.write_text(json.dumps(payload, indent=2), encoding="utf-8")
    print(f"Saved: {out_file}")

print("Done.")
```
